```python
import requests
from bs4 import BeautifulSoup
HEADERS = {"User-agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36"}
URL = "https://www.imdb.com/chart/boxoffice/?ref_=nv_ch_cht"
DATA = requests.get(URL, headers = HEADERS)
SOUP = BeautifulSoup(DATA.text, "html.parser")
period = SOUP.select_one("#boxoffice > h4").text

print(period)
print('-' * 20)

RANKS = 1
MOVIES = SOUP.select("#boxoffice > table > tbody > tr")

for movie in MOVIES:
    title = movie.select_one("td.titleColumn > a").text
    print("{0}. {1}".format(RANKS, title))
    RANKS += 1

```

```python
PAGE = 1
RANKS = 1
while PAGE < 6:

    URL = "https://www.newegg.com/d/Best-Sellers/Gaming/t/ID-8/Page-{}".format(PAGE)
    DATA = requests.get(URL, headers = HEADERS)
    SOUP = BeautifulSoup(DATA.text, "html.parser")

    product = SOUP.find_all("div", {"class" : "goods-info"})
    

    for item in product:
        title = item.find('a', {"class" : "goods-title"}).text

        if item.find('span', {"class" : "goods-price-value"}) == None:
            price = item.find('div', {"class" : "goods-price-map"}).text
        else:
            price = item.find('span', {"class" : "goods-price-value"}).text
    
        print("{0}. {1} ({2})".format(RANKS, title, price))
    
        RANKS += 1
    
    PAGE += 1

    
```
