```python
import pygame
import random

# initialize Pygame
pygame.init()

# set the width and height of the screen
screen_width = 800
screen_height = 600

# create the screen
screen = pygame.display.set_mode((screen_width, screen_height))

# set the caption of the screen
pygame.display.set_caption("My Game")

# define the colors
white = (255, 255, 255)
black = (0, 0, 0)
red = (255, 0, 0)
green = (0, 255, 0)

# set the font
font = pygame.font.Font(None, 50)

# define the function to draw text on the screen
def draw_text(text, color, x, y):
    text_surface = font.render(text, True, color)
    screen.blit(text_surface, (x, y))

# define the function to generate a random position
def generate_position():
    x = random.randint(0, screen_width)
    y = random.randint(0, screen_height)
    return x, y

# set the initial position of the player
player_x = screen_width // 2
player_y = screen_height // 2

# set the initial score to 0
score = 0

# set the clock
clock = pygame.time.Clock()

# set the game loop
running = True
while running:
    # handle events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:
                player_x -= 10
            elif event.key == pygame.K_RIGHT:
                player_x += 10
            elif event.key == pygame.K_UP:
                player_y -= 10
            elif event.key == pygame.K_DOWN:
                player_y += 10
                
    # generate a random position for the food
    food_x, food_y = generate_position()
    
    # check if the player has caught the food
    if abs(player_x - food_x) < 20 and abs(player_y - food_y) < 20:
        score += 1
        
    # fill the screen with white
    screen.fill(white)
    
    # draw the player
    pygame.draw.circle(screen, green, (player_x, player_y), 20)
    
    # draw the food
    pygame.draw.circle(screen, red, (food_x, food_y), 10)
    
    # draw the score
    draw_text("Score: {}".format(score), black, 10, 10)
    
    # update the screen
    pygame.display.update()
    
    # set the frame rate
    clock.tick(1)
    
# quit Pygame
pygame.quit()
```
